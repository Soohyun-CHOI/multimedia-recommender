```python
def clean_text(text):
  # removes contents in brackets and asterisks
  clean_text = re.sub(r"\[.*?\]", "", text)
  clean_text = re.sub(r"\*\*.*?\*\*", "", clean_text)

  # removes whitespaces before and after the string
  clean_text = clean_text.strip("\n")

  # replaced "\n" with a whitespace
  clean_text = clean_text.replace("\n", " ")

  # replace "(" and ")" with a whitespace
  clean_text = clean_text.replace("(", " ")
  clean_text = clean_text.replace(")", " ")

  # removes punctuations
  clean_text = re.sub(r"[^\w\s]", "", clean_text)

  # removes short words (for tokenization)
  clean_text = re.sub(r"\b\w{1,2}\b", "", clean_text)
  clean_text = re.sub(r"\s+", " ", clean_text).strip()

  # lowercase the text
  return clean_text.lower()
```

```python
def clean_categories(text):
  # removes brackets
  clean_text = text.replace("[", "")
  clean_text = clean_text.replace("]", "")

  # removes whitespaces before and after the string
  clean_text = clean_text.strip("\n")

  # replaced "\n" with a whitespace
  clean_text = clean_text.replace("\n", " ")

  # replace "(" and ")" with a whitespace
  clean_text = clean_text.replace("(", " ")
  clean_text = clean_text.replace(")", " ")

  # removes punctuations
  clean_text = re.sub(r"[^\w\s]", "", clean_text)

  # removes short words (for tokenization)
  clean_text = re.sub(r"\b\w{1,2}\b", "", clean_text)
  clean_text = re.sub(r"\s+", " ", clean_text).strip()

  # lowercase the text
  return clean_text.lower()
```

```python
import json

def get_json_data(input):
  #print(input)
  #print(re.search('''"*\'*"''', input))
  #print(re.search('''(["'])((?:\\\1|(?:(?!\1)).)*)(\1)''', input))

  #new_input = input.replace('"', '@').replace("'", '"').replace('@', "'").replace("[","").replace("]","").replace("None", '"None"')
  new_input = re.sub(r'"*\'*"', '"*"', input)
  new_input = new_input.replace("'", '"').replace("[","").replace("]","").replace("None", '"None"')
  #print(new_input)
  ret_list = []

  begin = new_input.rfind("{")
  while begin != -1:
    #print("Loop")
    #print(new_input[begin-1:].strip())
    try:
      if begin > 0:
        json_dict = json.loads(new_input[begin-1:].strip())
      else:
        json_dict = json.loads(new_input)
      #print(json_dict['name'])
      ret_list.append(json_dict['name'])
    except:
      ret_list.append("ERROR")
      #print("Unable to Parse: " + new_input[begin-1:].strip())

    end = new_input.rfind("}")
    new_input = new_input[:end]
    end = new_input.rfind("}")
    new_input = new_input[:end+1]
    begin = new_input.rfind("{")

  if len(ret_list) == 0:
    return None

  return ret_list
```
